```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#sns.set(style="darkgrid")
```

```python
df = pd.read_csv('D:/Datasets/avocado.csv').drop("Unnamed: 0", axis=1)
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df.Date)
```

```python
sns.lineplot(data=df, x='Date',y='AveragePrice')
```

```python
plt.figure(figsize=(20,9))
sns.lineplot(data=df, x='Date',y='AveragePrice')
plt.show()
```

```python
# Hue ile iki farklı kategorinin çizgi grafigi
```

```python
plt.figure(figsize=(20,9))
sns.lineplot(data = df, x='Date',y='AveragePrice', hue='type', ci=False)
plt.show()
```

```python
# region 'Albany','Atlanta','Boston' olanları al
df_cal = df[df['region'].isin(['Chicago','Boston'])]
```

```python
plt.figure(figsize=(20,9))
sns.lineplot(data = df_cal, x='Date',y='AveragePrice', hue='region', ci=False)
plt.show()
```
