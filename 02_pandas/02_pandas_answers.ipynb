```python
# simple_data.csv verisini kullanarak;

# 1. Sütun İsimlerini BÜYÜK HARFLİ yapınız
# 2. Yeni bir sütun içinde meslek,sehir sütunlarını aralarında çizgi olacak şekilde birleştiriniz.
# 3. aylik_gelir'i dolar cinsinden hesaplayarak aylik_gelir_dolar ismiyle yeni bir sütun ekleyiniz.
# 4. 3801'den başlayacak şekilde her bir satıra id numarası veriniz (yeni bir sütun içinde)
# 5. aylik_geliri baz alarak 0-2500 arasını düşük, 2500-7500 arasını orta ve 7500 üstünü yüksek olacak şekilde 
#    kategorilendirip gelir_kategori ismiyle yeni bir sütun olarak ekleyiniz.
```

```python
import pandas as pd
```

```python
df = pd.read_csv("D:/Datasets/simple_data.csv", sep=",")
```

```python
df.head()
```

```python
# 10.1. Sütun İsimlerini BÜYÜK HARFLİ yapınız
```

```python
df.columns
```

```python
#Herzaman elle düzeltecek kadar az nitelik olmayabilir. Bu tür işlemleri mümkün oldukça otomatize etmek gerekir.
new_columns = [i.upper() for i in df.columns]
print(new_columns)
```

```python
df.columns = new_columns
df.head(3)
```

```python
# 10.2. Yeni bir sütun içinde meslek, sehir sütunlarını aralarında çizgi olacak şekilde birleştiriniz.
```

```python
df['MESLEK_SEHIR'] = df.MESLEK + " - " + df.SEHIR
df.head(3)
```

```python
# 10.3. aylik_gelir'i dolar cinsinden hesaplayarak aylik_gelir_dolar ismiyle yeni bir sütun ekleyiniz.
```

```python
df['AYLIK_GELIR_DOLAR'] = df.AYLIK_GELIR / 5.9
df.head(3)
```

```python
# 10.4. 3801'den başlayacak şekilde her bir satıra id numarası veriniz (yeni bir sütun içinde)
```

```python
start = 3801
end = start + len(df)

df['ID'] = list(range(start,end))
```

```python
df.head(3)
```

```python
# 10.5. aylik_geliri baz alarak 0-2500 arasını düşük, 2500-7500 arasını orta ve 7500 üstünü yüksek olacak şekilde  kategorilendirip gelir_kategori ismiyle yeni bir sütun olarak ekleyiniz.
```

```python
max_deger = max(df.AYLIK_GELIR)
op_labels = ['düşük', 'orta', 'yüksek']
category = [0,2500,7500,max_deger+1000]
df['GELIR_LATEGORI'] = pd.cut(df['AYLIK_GELIR'], 
                              labels=op_labels, bins=category, 
                              include_lowest=False)
```

```python
df.head(30)
```

```python
# 10.5. Çözüm-2
```

```python
gelir_kategori = []
for i in df.AYLIK_GELIR:
    if i > 7500:
        gelir_kategori.append("yüksek")
    elif i > 2500:
        gelir_kategori.append("orta")
    else: gelir_kategori.append("düşük")
```

```python
len(gelir_kategori)
```

```python
df['GELIR_LATEGORI2'] = gelir_kategori
```

```python
df.head()
```

```python
# tips.csv dosyasını okuyarak günlere göre toplam ve ortalama bahşiş, ödeme tutarını hesaplayınız. 
# Sonuçları bahşiş ortalamasına göre azalan sıralayınız.
```

```python
tips = pd.read_csv("D:/Datasets/tips.csv")
tips.head()
```

```python
tips.groupby(['day'])[['tip','total_bill']] \
.agg(['mean','sum']) \
.sort_values([('tip', 'mean')], ascending=False) \
.head()
```

```python
# indeks resetleyerek
tips.groupby(['day'])[['tip','total_bill']] \
.agg(['mean','sum']) \
.reset_index() \
.sort_values([('tip', 'mean')], ascending=False) \
.head()
```
