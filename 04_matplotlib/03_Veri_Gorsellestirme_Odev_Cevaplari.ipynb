# 04_matplotlib: alıştırma - 4

```python
# OnlineRetail2.csv veri setini kullanarak günlere göre (Türkçe gün isimleriyle) ortalama satışları
# gösteren bar grafiği çiziniz.
```

```python
online_retail_df = pd.read_csv("D:\\Datasets\\OnlineRetail2.csv",
                 parse_dates=['InvoiceDate'], index_col="InvoiceDate")

online_retail_df['TotalPrice'] = online_retail_df['Quantity'] * online_retail_df['UnitPrice']
```

```python
online_retail_df.head()
```

```python
online_retail_df['day'] = online_retail_df.index.dayofweek
```

```python
# Günlern isimleri ile rakamları eşleştiren fonksiyon
def day_switch(i):
        switcher={
                
                0:'Pazartesi',
                1:'Salı',
                2:'Çarşamba',
                3:'Perşembe',
                4:'Cuma',
                5:'Cumartesi',
                6:'Pazar'
             }
        return switcher.get(i)
```

```python
day_switch(4)
```

```python
online_retail_df.head()
```

```python
day_names = []
for i in online_retail_df.day:
    day_names.append(day_switch(i))
```

```python
online_retail_df['day_names'] = day_names
```

```python
online_retail_df.head()
```

```python
online_retail_grp_days = online_retail_df \
.groupby(['day_names']).TotalPrice \
.agg(['mean']) \
.sort_values('mean',ascending=False) \
```

```python
online_retail_grp_days.head()
```

```python
# Generating the y positions. Later, we'll use them to replace them with labels.

x = online_retail_grp_days.index
y = online_retail_grp_days['mean']


y_pos = range(len(x))
plt.figure(figsize=(16,8)) 
plt.bar(x, y, align='center', alpha=0.7, color = 'g', width = 0.5) # alfa saydamlık derecesi
plt.xticks(y_pos, x, fontsize=18)
plt.xlabel('Günler', fontsize=18)
plt.xticks(rotation=90)
plt.ylabel('Satış Miktarı', fontsize=18)
plt.title('Günlere Göre Ortalama Satışlar', fontsize=22)
plt.show()
```
